```python
# ...existing code...
required_packages = packages_str.split(", ")

import importlib
import sys
import subprocess
import re

print("--- 패키지 확인 및 설치 시작 ---")
packages_for_version_check = []

def get_installed_version(pkg_name):
    try:
        import pkg_resources
        return pkg_resources.get_distribution(pkg_name).version
    except Exception:
        return None

for pkg_spec in required_packages:
    # 패키지 이름과 버전 분리 (== 연산자만 체크)
    match = re.match(r"([a-zA-Z0-9_\-]+)(==([\d\.]+))?", pkg_spec)
    if match:
        base_pkg_name = match.group(1)
        version_operator = match.group(2) or ""
        version_number = match.group(3) or ""
        packages_for_version_check.append(base_pkg_name)
    else:
        print(f"!!!! '{pkg_spec}' 에서 패키지 이름을 추출할 수 없습니다. 건너뜁니다.")
        continue

    need_install = False
    installed_version = get_installed_version(base_pkg_name)
    if installed_version is None:
        print(f"**** '{base_pkg_name}' 모듈을 찾을 수 없습니다. '{pkg_spec}' 설치를 시도합니다...")
        need_install = True
    elif version_operator == "==":
        # 버전 비교 필요
        if installed_version != version_number:
            print(f"**** '{base_pkg_name}' 버전이 {installed_version} 입니다. '{pkg_spec}'로 재설치합니다...")
            need_install = True

    if need_install:
        try:
            result = subprocess.run(
                [sys.executable, '-m', 'pip', 'install', pkg_spec, '-q'],
                check=True,
                capture_output=True,
                text=True
            )
            print(f"****** '{pkg_spec}' 설치 완료.")
        except subprocess.CalledProcessError as e:
            print(f"!!!!!!!! '{pkg_spec}' 설치 실패 !!!!!!!!")
            print(f"에러 메시지:\n{e.stderr}")
        except Exception as e:
            print(f"!!!!!!!! '{pkg_spec}' 설치 중 예상치 못한 에러 발생: {e}")
    else:
        print(f"******** '{base_pkg_name}' 모듈이 이미 설치되어 있고 버전도 맞습니다.")

print("\n--- version_information 로딩 ---")
try:
    get_ipython().run_line_magic("load_ext", "version_information")
except ImportError:
    print("!!!!!! 'version_information' 패키지가 설치되지 않았거나 로드할 수 없습니다.")
except Exception as e:
    print(f"!!!!!! version_information 로딩 중 에러 발생: {e}")

import time
now = time.strftime("%Y-%m-%d %H:%M:%S (%Z = GMT%z)")
print(f"\n이 노트북은 다음 시간에 생성되었습니다: {now} ")

print("\n--- 패키지 버전 확인 ---")
version_check_str = ", ".join(packages_for_version_check)
print(f"버전 확인 대상: {version_check_str}")

try:
    vv = get_ipython().run_line_magic("version_information", version_check_str)
    if hasattr(vv, 'packages'):
        print("\n설치된 패키지 버전 정보:")
        for i, pkg_info in enumerate(vv.packages):
            if isinstance(pkg_info, (list, tuple)) and len(pkg_info) >= 2:
                print(f"{i}: {pkg_info[0]:<20} {pkg_info[1]}")
            else:
                print(f"{i}: {pkg_info}")
    else:
        print("version_information 에서 패키지 정보를 가져오지 못했습니다.")
except NameError:
    print("!!!!!! 'version_information' extension이 로드되지 않았습니다.")
except Exception as e:
    print(f"!!!!!! %version_information 실행 중 에러 발생: {e}")

import warnings
warnings.filterwarnings("ignore")
# ...existing code...
```

```python
def check_Good_FITS(fits_path,  
                          fwhm_thresh=10, 
                          ellipticity_thresh=0.4, 
                          min_stars=5, 
                          verbose=True,
                          plot=True,
                          **kwargs):
    """
    가이드 불량 여부를 판단하는 함수.
    """
    msg = f"check Good FITS {fits_path.name}\n" 
    msg += f"fwhm_thresh : {fwhm_thresh}, " 
    msg += f"ellipticity_thresh : {ellipticity_thresh}, "
    msg += f"min_stars : {min_stars}, \n"

    with fits.open(fits_path) as hdul:
        data = hdul[0].data

        mean, median, std = sigma_clipped_stats(data, sigma=3.0)
        daofind = DAOStarFinder(fwhm=fwhm_thresh, 
                                threshold=5.0*std)
        sources = daofind(data - median)

        if sources is None :
            msg += f"No star found. \n"
            if verbose:
                print(f"{msg}")
            if plot == True :
                fig, axs = plt.subplots(1,1,
                            sharex=False, sharey=False, gridspec_kw=None)
                im = _astro_utilities.zimshow(axs, hdul[0].data, )
                axs.annotate(f'msg', fontsize=8,
                            xy=(0, 0), xytext=(10, -20), va='top', ha='left',
                            xycoords='axes fraction', textcoords='offset points')
                plt.show()
                plt.close()
            return False, msg
        elif len(sources) < min_stars:
            msg += f"Not enough stars({len(sources)}) to judge. \n"
            if verbose:
                print(f"{msg}")
            if plot == True :   
                fig, axs = plt.subplots(1, 1, 
                            sharex=False, sharey=False, gridspec_kw=None)
                im = _astro_utilities.zimshow(axs, hdul[0].data, )
                axs.annotate(f'msg', fontsize=8,
                            xy=(0, 0), xytext=(10, -20), va='top', ha='left',
                            xycoords='axes fraction', textcoords='offset points')
                plt.show()
                plt.close()
            return False, msg
        else:
            msg += f"Star count: {len(sources)}. \n"
            if verbose:
                print(f"{msg}")
            
            # Set up PSF model and fitter
            # psf_model = CircularGaussianPRF(x_0=0, y_0=0)
            # fitter = LevMarLSQFitter()
            # phot = PSFPhotometry(psf_model, fitter=fitter)

            # # Fit each detected source
            # for source in sources:
            #     cutout = Cutout2D(hdul[0].data, (source['xcentroid'], source['ycentroid']), size=20)
            #     result = phot(cutout)
            #     fwhm = result['fwhm']  # FWHM in pixels

            if plot == True :   
                fig, axs = plt.subplots(1, 1, 
                            sharex=False, sharey=False, gridspec_kw=None)
                im = _astro_utilities.zimshow(axs, hdul[0].data, )
                axs.annotate(f'msg', fontsize=8,
                            xy=(0, 0), xytext=(10, -20), va='top', ha='left',
                            xycoords='axes fraction', textcoords='offset points')
                plt.show()
                plt.close()
            return True, msg
```

```python
#%%
for DOINGDIR in DOINGDIRs[:] :
    DOINGDIR = Path(DOINGDIR)
    if verbose == True :
        print("DOINGDIR", DOINGDIR)

    # if _Python_utilities.check_string_in_file(log_file, DOINGDIR.name):
    #     print(f"'{DOINGDIR.name}' found in '{log_file}'")
    #     pass
    # else:
    # print(f"'{DOINGDIR.name}' not found in '{log_file}' or file does not exist.")

    BADFITSDIR = DOINGDIR / _astro_utilities.Bad_fits_dir
    if not BADFITSDIR.exists():
        os.makedirs("{}".format(str(BADFITSDIR)))
        if verbose == True :
            print("{} is created...".format(str(BADFITSDIR)))

    summary = yfu.make_summary(DOINGDIR/"*.fit*",
                                verify_fix=True,
                                ignore_missing_simple=True,
                                )
    if summary is not None :
        if verbose == True :
            print("len(summary):", len(summary))
            print("summary:", summary)
            #print(summary["file"][0])  
        df_light = summary.loc[summary["IMAGETYP"] == "LIGHT"].copy()
        df_light = df_light.reset_index(drop=True)
        if verbose == True :
            print("df_light:\n{}".format(df_light))
    if verbose == True :
        print("df_light :", df_light)

    for _, row  in df_light.iterrows():
        fpath = Path(row["file"])
        # fpath = Path(df_light["file"][1])
        if verbose == True :
            print("fpath :" ,fpath)
        # try :
        result, msg = check_Good_FITS(fpath,
                            fwhm_thresh=5, 
                            ellipticity_thresh=0.99, 
                            min_stars=10, 
                            plot=False,
                            verbose=True)
        if verbose == True :
            print("result: ", result)
        if result == False :
            shutil.move(fpath, BADFITSDIR/fpath.name)
            with open(BADFITSDIR/f"{fpath.stem}.log", "w") as f:
                f.write(msg + "\n")

            if verbose == True :
                print("#"*60)
                print("moved to: ", BADFITSDIR)
                print("#"*80)

        # except Exception as err :
        #     print("X"*60)
        #     _Python_utilities.write_log(err_log_file, f'''{fpath}, {str(err)}''', verbose=verbose)
        #     pass

    # _Python_utilities.write_log(log_?ose)
```
